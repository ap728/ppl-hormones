```python
# Setup - in this file we make plots
import matplotlib
import numpy as np
import scipy as sp
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import pandas as pd
import ggplot
%matplotlib inline
```

```python
# Import your cluster trace files

trace_1 = pd.read_csv("cluster_trace1.csv", header=0) #python starts counting at zero
trace_2 = pd.read_csv("cluster_trace2.csv", header=0) 
trace_3 = pd.read_csv("cluster_trace3.csv", header=0) 


```

```python
# Merge into one dataframe
trace_1["which"] = "Trace 1"
trace_2["which"] = "Trace 2"
trace_3["which"] = "Trace 3"

result = trace_1.append([trace_2, trace_3])
```

```python
# Plot trace

trace = ggplot.ggplot(ggplot.aes(x='t', y='cluster_num_count', color="which"), data=result)
trace = (trace + ggplot.geom_line()+ ggplot.theme_matplotlib() + 
ggplot.xlab("time where each point is a sample at n transitions"))

print(trace)
```
